# How Does a Resonator Measure a Qubit?

A superconducting qubit is usually coupled to a microwave resonator.

When the qubit and resonator are sufficiently detuned, they enter the dispersive regime. In this regime, the resonator frequency depends slightly on the state of the qubit.

As a result:

- the resonator frequency is shifted when the qubit is in $|g\rangle$,
- the resonator frequency is shifted in the opposite direction when the qubit is in $|e\rangle$.

By probing the resonator, we can therefore infer the state of the qubit without measuring the qubit directly.

Use the sliders to explore how the dispersive shift

$$
\chi \approx \frac{g^2}{\Delta}
$$

depends on:

- the coupling strength $g$,
- the detuning $\Delta$.

Questions:

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
```

```python
def lorentzian(f,f0,kappa):

    return 1-0.5/(1+((f-f0)/kappa)**2)
```

```python
def plot_dispersive(
    Delta=1000,
    g=100,
    fr=6000
):

    chi = g**2/Delta

    f = np.linspace(
        fr-20,
        fr+20,
        1000
    )

    plt.figure(figsize=(7,5))

    plt.plot(
        f,
        lorentzian(f,fr-chi,2),
        label="qubit |g>"
    )

    plt.plot(
        f,
        lorentzian(f,fr+chi,2),
        label="qubit |e>"
    )

    plt.xlabel("Frequency (MHz)")
    plt.ylabel("Response")

    plt.legend()
    plt.ylim(0,1)

    plt.title(
        rf"$\chi={chi:.2f}$ MHz"
    )

    plt.show()

interact(
    plot_dispersive,
    Delta=FloatSlider(
        value=1000,
        min=200,
        max=5000,
        step=50
    ),
    g=FloatSlider(
        value=100,
        min=10,
        max=300,
        step=5
    )
);
```
